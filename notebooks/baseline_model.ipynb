# Tokenization

```python
import spacy
from collections import Counter
import pandas as pd

# Load the spaCy English language model
nlp = spacy.load('en_core_web_sm')

def tokenize_and_count(df, text_column='Text'):
    # Tokenize each text in the specified column
    df['tokens'] = df[text_column].apply(lambda x: [token.lemma_.lower() for token in nlp(str(x))])
    ############ remove stop words
    
    # Flatten the list of lists into a single list of all tokens
    all_tokens = [token for sublist in df['tokens'] for token in sublist]

    # Count the occurrences of each token
    token_counts = Counter(all_tokens)

    # Create a word-to-index dictionary and vice versa
    word_to_index = {word: idx for idx, (word, _) in enumerate(token_counts.items(), start=1)}
    index_to_word = {idx: word for word, idx in word_to_index.items()}

    # Replace tokens with their corresponding IDs in the 'tokens' column
    df['ids'] = df['tokens'].apply(lambda tokens: [word_to_index[token] for token in tokens])

    return df, token_counts, word_to_index, index_to_word

# Example usage:
# Assuming you have a DataFrame 'df' with a 'Text' column
# df = pd.read_csv('your_dataset.csv')

# Call the function on the 'Text' column and save the results in the DataFrame
df, token_counts, word_to_index, index_to_word = tokenize_and_count(df, text_column='Text')

df
```

# Padding and Splitting

```python
import torch
from torch.utils.data import Dataset, DataLoader, random_split
from torch.nn.utils.rnn import pad_sequence

# torch.manual_seed(42)


class GFM_Dataset(Dataset):
    def __init__(self, df):
        self.data = df

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sequence = self.data.iloc[idx]['x']
        label = self.data.iloc[idx]['success']

        return {'sequence': torch.tensor(sequence, dtype=torch.int32), 'label': torch.tensor(label, dtype=torch.float32)}

def collate_fn(batch, max_seq_length):
    # Sort the batch based on sequence length (descending order)
    # batch.sort(key=lambda x: len(x['sequence']), reverse=True)
    
    # Truncate or pad sequences based on max_seq_length
    sequences = [item['sequence'][:max_seq_length] for item in batch]
    padded_sequences = pad_sequence(sequences, batch_first=True, padding_value=0)

    labels = torch.tensor([[item['label']] for item in batch])

    return padded_sequences, labels


# Convert sequences to lists of tensors
df['x'] = df['ids'].apply(lambda l: [int(i) for i in l])  

# Create the dataset
dataset = GFM_Dataset(df)

# Define the sizes of train, validation, and test sets
batch_size = 128
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size

# Split the dataset into train, validation, and test sets
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])

# Define max sequence length
max_seq_length = np.array([len(x) for x in df["ids"]]).max()  

# Define data loaders for each set
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, collate_fn=lambda batch: collate_fn(batch, max_seq_length))
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, collate_fn=lambda batch: collate_fn(batch, max_seq_length))
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, collate_fn=lambda batch: collate_fn(batch, max_seq_length))
```

# Model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchtext.vocab import GloVe
from torch.nn.utils.rnn import pack_padded_sequence, pad_packed_sequence
import torch.nn.functional as FF

# Define your sequence classification model
class BaselineModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, pretrained_embedding=None, freeze_embedding=True):
        super(BaselineModel, self).__init__()

        # Embedding layer
        if pretrained_embedding is not None:
            self.embedding = nn.Embedding.from_pretrained(pretrained_embedding, freeze=freeze_embedding)
        else:
            self.embedding = nn.Embedding(vocab_size, embedding_dim)

        # Linear layer
        self.linear1 = nn.Linear(embedding_dim, embedding_dim)
        self.dropout = nn.Dropout(0.5)
        self.linear2 = nn.Linear(embedding_dim, 1)
        
    def forward(self, x):
        embedded = self.embedding(x)
        # Assuming x is a tensor representing the sequence, you may need to adjust accordingly

        # Average pooling along the sequence dimension
        pooled = embedded.mean(dim=1)

        # Linear layer
        o = self.dropout(FF.relu(self.linear1(pooled)))
        output = FF.sigmoid(self.linear2(o))
        
        return output


vocab_size = len(word_to_index)  
embedding_dim = 300  

# Load pre-trained GloVe embedding (adjust the path accordingly)
glove = GloVe(name='6B', dim=embedding_dim)

# Create the model
model = BaselineModel(vocab_size, embedding_dim, pretrained_embedding=glove.vectors)

# Set the seed for reproducibility
torch.manual_seed(42)

# Example input sequence (replace with your actual input data)
input_sequence = torch.randint(0, vocab_size, (32, 10))  # Batch size of 32, sequence length of 10

# Forward pass
output = model(input_sequence)

# Print the output shape
print("Output Shape:", output.shape)
```

# Training

```python
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import torchmetrics.functional.classification as F

def train(model, train_loader, val_loader, criterion, optimizer, device, num_epochs=100, val_interval=10):
    
    model.train()
    losses = []
    accuracies = []
    f1_scores = []
    auc_scores = []
    val_losses = []
    val_accuracies = []
    val_f1_scores = []
    val_auc_scores = []
    
    for epoch in range(1, num_epochs+1):
        total_loss = 0.0
        accuracy = 0.0
        f1_score = 0.0
        auc_score = 0.0
        for x, y in train_loader:
            x, y = x.to(device), y.to(device)

            optimizer.zero_grad()
            outputs = model(x)
            loss = criterion(outputs, y)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            # Update Accuracy metric
            accuracy += F.binary_accuracy(outputs, y)
            f1_score += F.binary_f1_score(outputs, y)
            auc_score += F.binary_auroc(outputs, y.type(torch.int))
            
        average_loss = total_loss / len(train_loader)
        average_accuracy = accuracy / len(train_loader)
        average_f1 = f1_score / len(train_loader)
        average_auc = auc_score / len(train_loader)
        
        losses.append(average_loss)
        accuracies.append(average_accuracy.item())
        f1_scores.append(average_f1.item())
        auc_scores.append(average_auc.item())
        
        if epoch % val_interval == 0:
            val_loss, val_accuracy, val_f1, val_auc = validate(model, val_loader, criterion, device)
            val_losses.append(val_loss)
            val_accuracies.append(val_accuracy.item())
            val_f1_scores.append(val_f1.item())
            val_auc_scores.append(val_auc.item())
            
            print(f'Epoch {epoch}/{num_epochs},\
                  Training Loss: {average_loss:.4f}, Training Accuracy: {average_accuracy:.2%}\
                  Training F1: {average_f1:.4f}, Training AUC: {average_auc:.2%}\
                  Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.2%}\
                  Validation F1: {val_f1:.4f}, Validation AUC: {val_auc:.2%}')            
                  
        
        else:
            print(f'Epoch {epoch}/{num_epochs},\
                  Training Loss: {average_loss:.4f}, Training Accuracy: {average_accuracy:.2%}\
                  Training F1: {average_f1:.4f}, Training AUC: {average_auc:.2%}')

    
    epochs = range(1, len(losses) + 1)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    # Plot training and validation losses
    ax1.plot(epochs, losses, label='Training Loss')
    ax1.plot(range(val_interval, len(val_losses) * val_interval + 1, val_interval), val_losses, label='Validation Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.grid(True)
    ax1.legend()

    # Plot validation accuracy
    ax2.plot(epochs, accuracies, label='Training Accuracy')
    ax2.plot(range(val_interval, len(val_accuracies) * val_interval + 1, val_interval), val_accuracies, label='Validation Accuracy', marker='o')
    ax2.plot(epochs, f1_scores, label='Training F1')
    ax2.plot(range(val_interval, len(val_accuracies) * val_interval + 1, val_interval), val_f1_scores, label='Validation F1', marker='o')
    ax2.plot(epochs, auc_scores, label='Training AUC')
    ax2.plot(range(val_interval, len(val_accuracies) * val_interval + 1, val_interval), val_auc_scores, label='Validation AUC', marker='o')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('%')
    ax2.grid(True)
    ax2.legend()

    plt.tight_layout()
    plt.show()

def validate(model, val_loader, criterion, device):
    model.eval()
    total_loss = 0.0
    accuracy = 0.0
    f1_score = 0.0
    auc_score = 0.0
    
    with torch.no_grad():
        for x, y in val_loader:
            x, y = x.to(device), y.to(device)
            
            outputs = model(x)
            loss = criterion(outputs, y)
            total_loss += loss.item()
            # Update Accuracy metric
            accuracy += F.binary_accuracy(outputs, y)
            f1_score += F.binary_f1_score(outputs, y)
            auc_score += F.binary_auroc(outputs, y.type(torch.int))
            
        average_loss = total_loss / len(val_loader)
        average_accuracy = accuracy / len(val_loader)
        average_f1 = f1_score / len(val_loader)
        average_auc = auc_score / len(val_loader)
    return average_loss, average_accuracy, average_f1, average_auc

# Example usage:
# Assuming you have a DataLoader 'train_loader' for training and 'val_loader' for validation
# Define your model, optimizer, and loss criterion
vocab_size = len(word_to_index)  
embedding_dim = 300  
lr = 1e-3

model = BaselineModel(vocab_size, embedding_dim, pretrained_embedding=glove.vectors)
optimizer = optim.AdamW(model.parameters(), lr=lr)
criterion = nn.BCELoss()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
train(model, train_loader, val_loader, criterion, optimizer, device, num_epochs=1000, val_interval=10)
```
